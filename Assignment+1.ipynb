### Question 4

```python
def answer_four():
    #G_df = pd.read_csv('Employee_Movie_Choices.txt', delim_whitespace=True, 
                   #header=None, names=['employee', 'movie'])
    #G_m = nx.from_pandas_dataframe(G_df, 'employee', 'movie')
    #G_m.add_nodes_from(employee, bipartite = 0, type='employee')
    #G_m.add_nodes_from(movie, bipartite = 1, type='movie')
    G_m = nx.read_edgelist("Employee_Movie_Choices.txt", delimiter='\t')
    G_m.add_nodes_from(['Pablo',
                 'Lee',
                 'Georgia',
                 'Vincent',
                 'Andy',
                 'Frida',
                 'Joan',
                 'Claude'], bipartite=0)
    
    G_m.add_nodes_from(['The Shawshank Redemption',
              'Forrest Gump',
              'The Matrix',
              'Anaconda',
              'The Social Network',
              'The Godfather',
              'Monty Python and the Holy Grail',
              'Snakes on a Plane',
              'Kung Fu Panda',
              'The Dark Knight',
              'Mean Girls'], bipartite=1)
    G_m.add_node('Pablo',type='employee') 
    G_m.add_node('Lee', type='employee') 
    G_m.add_node('Georgia', type='employee') 
    G_m.add_node('Vincent', type='employee') 
    G_m.add_node('Andy', type='employee') 
    G_m.add_node('Frida', type='employee') 
    G_m.add_node('Joan', type='employee')
    G_m.add_node('Claude', type='employee') 
    
    G_m.add_node('The Shawshank Redemption', type='movie') 
    G_m.add_node('Forrest Gump', type='movie') 
    G_m.add_node('The Matrix', type='movie') 
    G_m.add_node('Anaconda', type='movie') 
    G_m.add_node('The Social Network', type='movie') 
    G_m.add_node('The Godfather', type='movie') 
    G_m.add_node('Monty Python and the Holy Grail', type='movie') 
    G_m.add_node('Snakes on a Plane', type='movie') 
    G_m.add_node('Kung Fu Panda', type='movie') 
    G_m.add_node('The Dark Knight', type='movie') 
    G_m.add_node('Mean Girls', type='movie') 
    
    X = ['Pablo','Lee','Georgia', 'Vincent','Andy','Frida','Joan', 'Claude']
    B_m = bipartite.weighted_projected_graph(G_m,X) 
    df_mw = pd.DataFrame(B_m.edges(data=True), columns=['employee1', 'employee2', 'weight'])
    df_mw['weight'] = df_mw['weight'].map(lambda x: x['weight'])
    df_mw2 = df_mw.copy()
    df_mw2.rename(columns={'employee1': 'employee2', 'employee2': 'employee1'}, inplace=True)
    df_common_movies = pd.concat([df_mw, df_mw2], ignore_index=True)
    df_common_movies.sort_values(by = 'employee1') 
    
    df_r = pd.read_csv('Employee_Relationships.txt', delim_whitespace=True, 
                   header=None, names=['employee1', 'employee2', 'r'])
    
    #df_join = df_r.join(df_mw, )
    df_merged = pd.merge(df_r, df_common_movies, how='left')
    df_merged['weight'].fillna(value=0, inplace=True)
    correlation_score = df_merged['weight'].corr(df_merged['r'])
    
    # Your Code Here
    # Your Code Here
    
    return correlation_score #correlation_score#G_r.edges(data=True)# Your Answer Here
answer_four()
```
